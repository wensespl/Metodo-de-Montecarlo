# Método de Simpson

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def funcion_simple(x):
    return((3)*(x**2)) #Función Polinómica

def simps(f,a,b,N):
    '''Aproximación de la Integral usando la Regla de Simpson.
    
    Parametros
    ----------
    f : Función de una sola variable
    a , b : Límites de integración [a,b]
    N : Númbero de subintervalos de [a,b]. Debe ser par.

    Retornará
    ---------
    Una aproximación de la integral de la función usando la Regla de Simpson.
    
    '''
    if N % 2 == 1:
        raise ValueError("N debe ser un número par.")
    dx = (b-a)/N
    x = np.linspace(a,b,N+1)
    y = f(x)
    S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S

X=np.linspace(-20,20,1000)
plt.plot(X,funcion_simple(X)) #Muestra la Función Polinómica
plt.show()

simps(lambda x : 3*x**2,0,1,1000) #Muestra el resultado de la integral de la función con N = 1000
```
